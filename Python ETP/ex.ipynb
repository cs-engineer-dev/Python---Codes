```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# Replace 'sales_data.csv' with the path to your dataset file
df = pd.read_csv("C:/Users/hp/Downloads/My_Work/Sales_Data.csv")

# Display the first few rows of the dataset
print("Dataset Head:")
print(df.head())

# 1. Handling missing values: Replace missing values in 'Sales' column with the column mean
if 'Sales' in df.columns:
    mean_sales = df['Sales'].mean()
    df['Sales'] = df['Sales'].fillna(mean_sales)
    print(f"\nMissing values in 'Sales' column replaced with mean: {mean_sales}")

# 2. Finding maximum and minimum sales
max_sales = df['Sales'].max()
min_sales = df['Sales'].min()
print(f"\nMaximum Sales: {max_sales}")
print(f"Minimum Sales: {min_sales}")

# 3. Filtering: Keep products with sales greater than 200
filtered_df = df[df['Sales'] > 200]
print("\nFiltered DataFrame (Sales > 200):")
print(filtered_df)

# 4. Sorting products by sales in descending order
sorted_df = df.sort_values(by='Sales', ascending=False)
print("\nDataFrame Sorted by Sales:")
print(sorted_df)

# 5. Bar graph for product sales
plt.figure(figsize=(10, 6))
plt.bar(sorted_df['Product'], sorted_df['Sales'], color='skyblue')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Bar Graph')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
