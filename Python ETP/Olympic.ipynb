```python
medal_data = {
    '2020':[{'event': 'Hockey','athlete':'Belgium Men.s Team','medal':'Gold','country': 'Belgium'},
            {'event': 'Athletics', 'athlete': 'Naffisatou Thiam', 'medal': 'Gold', 'country': 'Belgium'},
            {'event': 'Cycling', 'athlete': 'Wout van Aert', 'medal': 'Silver', 'country': 'Belgium'}],
    '2016':[{'event': 'Athletics','athlete':'Naffisatou Thiam','medal':'Gold','country': 'Belgium'},
            {'event': 'Hockey', 'athlete': 'Belgium Men.s Team', 'medal': 'Silver', 'country': 'Belgium'}],
    '2012':[{'event': 'Cycling','athlete':'Gijs Van Hoecke','medal':'Bronze','country': 'Belgium'},
            {'event': 'Judo', 'athlete': 'Charline van Snick', 'medal': 'Bronze', 'country': 'Belgium'}],
    '2008':[{'event': 'Cycling','athlete':'Greg Van Avermaet','medal':'Gold','country': 'Belgium'}],
    '2004':[{'event': 'Tennis','athlete':'Justine Henin','medal':'Gold','country': 'Belgium'}]
}

def add_details(year, event, athlete, medal, country):
    year = str(year).strip()
    event = str(event).strip()
    athlete = str(athlete).strip()
    medal = str(medal).strip().title()
    country = str(country).strip()
    medal_data.setdefault(year, []).append({
        'event': event,
        'athlete': athlete,
        'medal': medal,
        'country': country
    })
    print(f"Details added for {athlete} in {year}.")

def view_details(year):
    year = str(year).strip()
    if year in medal_data:
        for record in medal_data[year]:
            print(f"{record['event'].title()}: {record['athlete']} won {record['medal']} for {record['country']}.")
    else:
        print(f"Data is not available for year: {year}.")

def view_medal_count_by_country(country):
    country = str(country).strip()
    count = 0
    for year_records in medal_data.values():
        for record in year_records:
            if record.get('country', '').lower() == country.lower():
                count += 1
    print(f"{country} has won {count} medal(s) in the dataset.")

def search_athlete_performance(athlete):
    athlete_q = str(athlete).strip()
    found = False
    for year, year_records in medal_data.items():
        for record in year_records:
            if record.get('athlete', '').lower() == athlete_q.lower():
                print(f"{record['athlete']} won {record['medal']} in {record['event']} in {year}.")
                found = True
    if not found:
        print(f"Data is not available for this athlete: {athlete_q}.")

def view_event_details(event):
    event_q = str(event).strip()
    found = False
    for year, year_records in medal_data.items():
        for record in year_records:
            if record.get('event', '').lower() == event_q.lower():
                print(f"In {year}, {record['athlete']} won {record['medal']} for {record['country']} in {record['event']}.")
                found = True
    if not found:
        print(f"Data is not available for this event: {event_q}.")

def main():
    while True:
        print("\n-----------------Olympic Data Explorer------------------")
        print("Menu:")
        print("1. Add Details")
        print("2. View Details by Year")
        print("3. View Medal Count by Country")
        print("4. Search Athlete Performance")
        print("5. View Event Details")
        print("6. Exit")

        choice = input("Enter Your Choice: ").strip()
        if choice == '1':
            year = input("Enter the year: ").strip()
            event = input("Enter the event: ").strip()
            athlete = input("Enter the athlete's name: ").strip()
            medal = input("Enter the medal (Gold/Silver/Bronze): ").strip()
            country = input("Enter the Country name: ").strip()
            add_details(year, event, athlete, medal, country)

        elif choice == '2':
            year = input("Enter the year to view details: ").strip()
            view_details(year)

        elif choice == '3':
            country = input("Enter the Country name to view medal count: ").strip()
            view_medal_count_by_country(country)

        elif choice == '4':
            athlete = input("Enter the athlete name to view performance: ").strip()
            search_athlete_performance(athlete)

        elif choice == '5':
            event = input("Enter the event to view details: ").strip()
            view_event_details(event)

        elif choice == '6':
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
